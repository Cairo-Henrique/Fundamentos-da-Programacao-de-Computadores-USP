```python
import numpy as np
import time
import matplotlib.pyplot as plt

def random_linear_combination(n):

  a = np.random.rand(n)
  b = np.random.rand(n)

  alfa = np.random.rand()
  beta = np.random.rand()

  c = alfa * a + beta * b

  return c

def random_power_matrix(n, m):

  A = np.random.rand(n, n)

  return np.linalg.matrix_power(A, m)

# valores de n
n_random_linear_combination = np.array([10**4, 10**5, 10**6, 10**7])
n_random_power_matrix = np.array([500, 1000, 1500])

# valores de tempo
time_random_linear_combination = np.array([])
time_random_power_matrix_m2 = np.array([])
time_random_power_matrix_m3 = np.array([])
time_random_power_matrix_m4 = np.array([])

# medir tempo
for n in n_random_linear_combination:
  start = time.time()
  c = random_linear_combination(n)
  end = time.time()
  t = end - start
  time_random_linear_combination = np.append(time_random_linear_combination, t)

m = 2
for n in n_random_power_matrix:
  start = time.time()
  c = random_power_matrix(n, m)
  end = time.time()
  t = end - start
  time_random_power_matrix_m2 = np.append(time_random_power_matrix_m2, t)

m = 3
for n in n_random_power_matrix:
  start = time.time()
  c = random_power_matrix(n, m)
  end = time.time()
  t = end - start
  time_random_power_matrix_m3 = np.append(time_random_power_matrix_m3, t)

m = 4
for n in n_random_power_matrix:
  start = time.time()
  c = random_power_matrix(n, m)
  end = time.time()
  t = end - start
  time_random_power_matrix_m4 = np.append(time_random_power_matrix_m4, t)

# Gr√°fico 1: random_linear_combination
plt.figure()
plt.plot(n_random_linear_combination, time_random_linear_combination, marker="o", label="random_linear_combination")
plt.xlabel("n")
plt.ylabel("Tempo (s)")
plt.title("Tempo de execu√ß√£o - random_linear_combination")
plt.legend()
plt.grid(True)
plt.show()

plt.figure()
plt.loglog(n_random_linear_combination, time_random_linear_combination, marker="o", label="random_linear_combination")
plt.xlabel("n")
plt.ylabel("Tempo (s)")
plt.title("Tempo de execu√ß√£o - random_linear_combination")
plt.legend()
plt.grid(True, which="both")  # mostra grid em escala log
plt.show()

# Gr√°fico 2: random_power_matrix
plt.figure()
plt.plot(n_random_power_matrix, time_random_power_matrix_m2, marker="o", color="red", label="random_power_matrix (m=2)")
plt.plot(n_random_power_matrix, time_random_power_matrix_m3, marker="o", color="green", label="random_power_matrix (m=3)")
plt.plot(n_random_power_matrix, time_random_power_matrix_m4, marker="o", color="blue", label="random_power_matrix (m=4)")
plt.xlabel("n")
plt.ylabel("Tempo (s)")
plt.title("Tempo de execu√ß√£o - random_power_matrix")
plt.legend()
plt.grid(True)
plt.show()

plt.figure()
plt.loglog(n_random_power_matrix, time_random_power_matrix_m2, marker="o", color="red", label="random_power_matrix (m=2)")
plt.loglog(n_random_power_matrix, time_random_power_matrix_m3, marker="o", color="green", label="random_power_matrix (m=3)")
plt.loglog(n_random_power_matrix, time_random_power_matrix_m4, marker="o", color="blue", label="random_power_matrix (m=4)")
plt.xlabel("n")
plt.ylabel("Tempo (s)")
plt.title("Tempo de execu√ß√£o - random_power_matrix")
plt.legend()
plt.grid(True, which="both")  # ativa grid para log scale
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def random_matrix_multiplication(A, B):

    m, r = A.shape
    r_b, n = B.shape

    # Checar dimens√µes
    if r != r_b:
        raise ValueError("Dimens√µes erradas!")

    C = np.zeros((m, n))

    # calcular produto matricial manualmente
    for i in range(m):
        for j in range(n):
            for k in range(r):
                C[i, j] += A[i, k] * B[k, j]

    return C

dimensoes = np.array([50, 100, 150, 200, 250])
time_random_matrix_multiplication = np.array([])
time_arrouba = np.array([])

for dim in dimensoes:
    A = np.random.rand(dim, dim)
    B = np.random.rand(dim, dim)

    start = time.time()
    C = random_matrix_multiplication(A, B)
    end = time.time()
    t = end - start
    time_random_matrix_multiplication = np.append(time_random_matrix_multiplication, t)

    start = time.time()
    C = A @ B
    end = time.time()
    t = end - start
    time_arrouba = np.append(time_arrouba, t)

plt.figure()
plt.plot(dimensoes, time_random_matrix_multiplication, marker="o", label="random_matrix_multiplication")
plt.plot(dimensoes, time_arrouba, marker="o", label="A @ B")
plt.xlabel("Dimens√£o")
plt.ylabel("Tempo (s)")
plt.title("Tempo de execu√ß√£o - escala linear")
plt.legend()
plt.grid(True)
plt.show()

plt.figure()
plt.loglog(dimensoes, time_random_matrix_multiplication, marker="o", label="random_matrix_multiplication")
plt.plot(dimensoes, time_arrouba, marker="o", label="A @ B")
plt.xlabel("Dimens√£o")
plt.ylabel("Tempo (s)")
plt.title("Tempo de execu√ß√£o - escala loglog")
plt.legend()
plt.grid(True, which="both")  # mostra grid em escala log
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def series_pi(n):
  pi_n = 0
  for i in range(n):
    pi_n += (-1)**i / (2 * i + 1)
  return 4 * pi_n

def erro_series_pi(n):
  pi_n = series_pi(n)
  return abs(np.pi - pi_n)

valores_n = np.arange(1, 1000)

valores_erro = np.array([erro_series_pi(n) for n in valores_n])

# Gr√°fico

plt.figure
plt.plot(valores_n, valores_erro, label="abs(pi - pi_n)")
plt.xlabel("n")
plt.ylabel("Valor")
plt.title("Erro de pi - escala linear")
plt.legend()
plt.grid(True)
plt.show()

plt.figure
plt.loglog(valores_n, valores_erro, label="abs(pi - pi_n)")
plt.xlabel("n")
plt.ylabel("Valor")
plt.title("Erro de pi - escala loglog")
plt.legend()
plt.grid(True, which="both")  # mostra grid em escala log
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 5000
x_0 = 0.1

def sequencia(N, x_0, a):
  sequencia = np.array([x_0])
  for i in range(1, N):
    x = a * sequencia[i-1] * (1 - sequencia[i-1])
    sequencia = np.append(sequencia, x)
  return sequencia

def mean(sequencia):
  soma = 0
  for x in sequencia:
    soma += x
  return soma / sequencia.size

def var(sequencia):
  soma = 0
  media = mean(sequencia)
  for x in sequencia:
    soma += (x - media)**2
  return soma / (sequencia.size - 1)

valores_a = (4*np.random.rand(5)).round(1)

for a in valores_a:
  print(f"a = {a}")
  sequencia_a = sequencia(N, x_0, a)
  print(f"M√©dia manual: {mean(sequencia_a)}")
  print(f"M√©dia numpy: {np.mean(sequencia_a)}")
  print(f"Vari√¢ncia manual: {var(sequencia_a)}")
  print(f"Vari√¢ncia numpy: {np.var(sequencia_a)}")
```

```python
# Considerando que n√£o √© para contar repeti√ß√µes de caracteres

def funcao(string1, string2):
  tamanho1 = len(string1)
  tamanho2 = len(string2)

  char_iguais = 0
  char_ja_comparados = [] # Lista para acompanhar quais caracteres j√° foram checados

  for char in string1:
    if char not in char_ja_comparados:
      if char in string2:
        char_ja_comparados.append(char)
        char_iguais += 1

  return tamanho1, tamanho2, char_iguais

print(funcao('cairo','ana'))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import sin, cos, tan

g = 9.81 # acelera√ß√£o gravitacional em m/s¬≤

entradas = []

while True:
  entrada = input('Digite o √¢ngulo e a velocidade inicial separados por espa√ßo: ')
  if entrada == '': # Apertar enter para parar
    break
  entrada = [float(i) for i in entrada.split()] # Formatar entrada como lista de dois elementos float
  entradas.append(entrada)

#print(entradas)

def f(posicao, angulo, velocidade): # par√°bola

  y = posicao * tan(angulo) - (g * posicao ** 2) / (2 * velocidade ** 2 * cos(angulo) ** 2)

  return y

# Gr√°fico
plt.figure()

for entrada in entradas:
  angulo = entrada[0] * np.pi / 180 # converter para radianos
  angulo_graus = entrada[0]
  velocidade = entrada[1]

  alcance_max = velocidade**2 * sin(2*angulo) / g

  posicao = np.linspace(0, alcance_max, 100)

  plt.plot(posicao, f(posicao, angulo, velocidade), label = f'√¢ngulo = {angulo_graus}, v_0 = {velocidade}')

plt.xlabel("Posi√ß√£o (m)")
plt.ylabel("Altura (m)")
plt.title("Gr√°fico do lan√ßamento obl√≠quo")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# par√¢metros
n_steps = 1000
n_trajectories = 10
delta = 1.0

# retorna arrays de x e y
def random_walk_2d(n_steps, delta):
    x = np.zeros(n_steps + 1)
    y = np.zeros(n_steps + 1)
    for i in range(1, n_steps + 1):
        d = np.random.randint(4)  # 0,1,2,3 igualmente prov√°veis
        if d == 0:      # direita
            x[i] = x[i-1] + delta
            y[i] = y[i-1]
        elif d == 1:    # esquerda
            x[i] = x[i-1] - delta
            y[i] = y[i-1]
        elif d == 2:    # cima
            x[i] = x[i-1]
            y[i] = y[i-1] + delta
        else:           # baixo (d == 3)
            x[i] = x[i-1]
            y[i] = y[i-1] - delta
    return x, y

plt.figure()
for i in range(n_trajectories):
    x, y = random_walk_2d(n_steps, delta)
    plt.plot(x, y, linewidth=0.8)

plt.scatter([0], [0], color='green', label='origem')
plt.axis('equal')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Trajet√≥ria aleat√≥ria 2D ‚Äî {n_trajectories} trajet√≥rias de {n_steps} passos')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Conectar dois n√≥s com outro n√≥
def connect(i, j, edges):
  edges.append((i,j))
  return edges

# Deletar uma aresta
def delete_edge(i, j, coords, edges):
  if (i, j) in edges:
    edges.remove((i, j))
  if (j, i) in edges:
    edges.remove((j, i)) # remove o par i,j e j,i
  if i not in [k[0] for k in edges] and i not in [k[1] for k in edges]: # se i n√£o tiver mais nenhuma aresta
    coords = np.delete(coords, i, axis=0) # deleta a linha contendo i da matriz de n√≥s
    edges = [(a-1 if a > i else a, b-1 if b > i else b) for a, b in edges] # atualiza os nomes dos n√≥s a partir de i
  if j not in [k[0] for k in edges] and j not in [k[1] for k in edges]: # se j n√£o tiver mais nenhuma aresta
    coords = np.delete(coords, j, axis=0) # deleta a linha contendo j da matriz de n√≥s
    edges = [(a-1 if a > j else a, b-1 if b > j else b) for a, b in edges] # atualiza os nomes dos n√≥s a partir de i
  return coords, edges

def delete_node(i, coords, edges):
  coords = np.delete(coords, i, axis=0) # deleta a linha contendo i da matriz de n√≥s
  for aresta in edges: # remover todas as arestas contendo i
    if i in aresta:
      edges.remove(aresta)
  edges = [(a-1 if a > i else a, b-1 if b > i else b) for a, b in edges] # atualiza os nomes dos n√≥s a partir de i
  return coords, edges

# coordenadas dos n√≥s
coords = np.array([
    [0.0, 0.0],  # n√≥ 0
    [2.0, 0.0],  # n√≥ 1
    [2.0, 2.0],  # n√≥ 2
    [4.0, 0.0],  # n√≥ 3
    [4.0, 2.0],  # n√≥ 4
    [1.0, 5.0],  # n√≥ 5
    [2.0, 4.0],  # n√≥ 6
    [4.0, 4.0],  # n√≥ 7
    [2.0, 6.0],  # n√≥ 8
    [4.0, 6.0]   # n√≥ 9
])

edges = [(0,1),(1,2),(1,3),(2,4),(3,4),(2,6),(5,6),(6,7),(5,7),(7,8),(8,9)]

# Fun√ß√µes aplicadas
#edges = connect(0, 6, edges)
#coords, edges = delete_edge(8, 9, coords, edges)
#coords, edges = delete_node(3, coords, edges)

# Criar matriz de adjac√™ncia (NxN)
N = len(coords)
A = np.zeros((N, N))
for i,j in edges:
    A[i,j] = 1
    A[j,i] = 1 # A = A^T

# Gr√°fico

plt.figure(figsize=(10,10))
# desenha arestas
for i in range(N): # para cada n√≥ i
    for j in range(i+1, N): # para cada n√≥ j ap√≥s i. Isso evita repeti√ß√£o de arestas
        if A[i,j] == 1:
            xi, yi = coords[i] # coordenadas n√≥ i
            xj, yj = coords[j] # coordenadas n√≥ j
            plt.plot([xi, xj], [yi, yj], linewidth=1, color='black') # conecta i e j por plot()

# desenha n√≥s e r√≥tulos
valores_x = [i[0] for i in coords]
valores_y = [i[1] for i in coords]
plt.scatter(valores_x, valores_y)
for k,(x,y) in enumerate(coords):
    plt.text(x+0.08, y+0.08, str(k), fontsize=10)

plt.title("Grafo")
plt.xlabel("x")
plt.ylabel("y")
plt.axis("equal")
plt.grid(False)
plt.show()
```
